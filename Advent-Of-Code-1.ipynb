```python
import pandas as pd
import numpy as np

def display_data(day, n):
    with open(f'data/day{day}.txt') as f:
        for _ in range(n):
            print(f.readline(), end='')
```

```python
display_data(1, 10)
```

```python
df = pd.read_csv('data/day1.txt', names=['calories'], skip_blank_lines=False) # Keep the blank lines as NaN
```

```python
df
```

```python
df['flag'] = np.where(np.isnan(df['calories']),1,0)
```

```python
df['elf'] = df["flag"].diff().ne(0).cumsum()
```

```python
df.dropna()
```

```python
df
```

```python
df.groupby(['elf']).sum().sort_values('calories', ascending=False)
```

```python
df.groupby(['elf']).sum().nlargest(1,'calories') # part 1
```

```python
df.groupby(['elf']).sum().nlargest(3,'calories').sum() # part 2
```
